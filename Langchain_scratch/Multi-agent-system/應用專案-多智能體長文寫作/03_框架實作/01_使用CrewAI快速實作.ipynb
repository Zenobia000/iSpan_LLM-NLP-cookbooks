## 2. åŸºæœ¬æ¦‚å¿µï¼šAgentï¼ˆæ™ºèƒ½é«”ï¼‰å’Œ Taskï¼ˆä»»å‹™ï¼‰

```python
from crewai import Agent, Task, Crew
import os
from dotenv import load_dotenv

load_dotenv()

# å‰µå»ºä¸€å€‹ç°¡å–®çš„æ™ºèƒ½é«”
researcher = Agent(
    role='ç ”ç©¶å“¡',
    goal='ç ”ç©¶å’Œæ”¶é›†è³‡è¨Š',
    backstory='ä½ æ˜¯ä¸€ä½ç¶“é©—è±å¯Œçš„ç ”ç©¶å“¡ï¼Œæ“…é•·æ‰¾åˆ°æº–ç¢ºçš„è³‡è¨Šã€‚',
    verbose=True  # é¡¯ç¤ºæ€è€ƒéç¨‹
)

print("âœ… ç ”ç©¶å“¡æ™ºèƒ½é«”å‰µå»ºå®Œæˆï¼")
```

```python
# å‰µå»ºä¸€å€‹ä»»å‹™
research_task = Task(
    description='ç ”ç©¶é—œæ–¼ã€Œäººå·¥æ™ºæ…§åœ¨æ•™è‚²çš„æ‡‰ç”¨ã€çš„è³‡è¨Š',
    expected_output='æä¾› 3 å€‹é‡è¦ç™¼ç¾ï¼Œæ¯å€‹ç”¨ä¸€æ®µè©±èªªæ˜',
    agent=researcher
)

print("âœ… ç ”ç©¶ä»»å‹™å‰µå»ºå®Œæˆï¼")
```

## 3. å‰µå»ºåœ˜éšŠï¼ˆCrewï¼‰

```python
# å‰µå»ºåœ˜éšŠä¸¦åŸ·è¡Œä»»å‹™
crew = Crew(
    agents=[researcher],
    tasks=[research_task],
    verbose=True
)

# åŸ·è¡Œï¼
result = crew.kickoff()
print("\n" + "="*50)
print("æœ€çµ‚çµæœï¼š")
print("="*50)
print(result)
```

## 4. å»ºç«‹å®Œæ•´çš„å¯«ä½œåœ˜éšŠ

```python
# å‰µå»ºå¤šå€‹å°ˆæ¥­æ™ºèƒ½é«”

# ç ”ç©¶å“¡
researcher = Agent(
    role='ç ”ç©¶å°ˆå®¶',
    goal='æ·±å…¥ç ”ç©¶ä¸»é¡Œï¼Œæ‰¾å‡ºé—œéµè³‡è¨Š',
    backstory='ä½ æ˜¯è³‡æ·±ç ”ç©¶å“¡ï¼Œæ“…é•·å¾å¤šè§’åº¦åˆ†æå•é¡Œ',
    verbose=True
)

# å¯«ä½œè€…
writer = Agent(
    role='å…§å®¹å¯«ä½œå°ˆå®¶',
    goal='å°‡ç ”ç©¶è³‡æ–™è½‰åŒ–ç‚ºå„ªè³ªæ–‡ç« ',
    backstory='ä½ æ˜¯å°ˆæ¥­ä½œå®¶ï¼Œæ“…é•·å°‡è¤‡é›œæ¦‚å¿µç”¨ç°¡å–®èªè¨€è¡¨é”',
    verbose=True
)

# ç·¨è¼¯
editor = Agent(
    role='ç·¨è¼¯',
    goal='ç¢ºä¿æ–‡ç« å“è³ªå’Œæº–ç¢ºæ€§',
    backstory='ä½ æ˜¯è³‡æ·±ç·¨è¼¯ï¼Œæ³¨é‡æ–‡ç« çš„é‚è¼¯æ€§å’Œå¯è®€æ€§',
    verbose=True
)

print("âœ… ä¸‰å€‹å°ˆæ¥­æ™ºèƒ½é«”å‰µå»ºå®Œæˆï¼")
```

```python
# å®šç¾©ä»»å‹™æµç¨‹

# ä»»å‹™1ï¼šç ”ç©¶
research_task = Task(
    description="""
    ç ”ç©¶ä¸»é¡Œï¼šæ°¸çºŒç™¼å±•èˆ‡ä¼æ¥­è²¬ä»»
    
    è«‹å¾ä»¥ä¸‹è§’åº¦ç ”ç©¶ï¼š
    1. ä»€éº¼æ˜¯æ°¸çºŒç™¼å±•ï¼Ÿ
    2. ä¼æ¥­å¦‚ä½•å¯¦è¸æ°¸çºŒç™¼å±•ï¼Ÿ
    3. æˆåŠŸæ¡ˆä¾‹æœ‰å“ªäº›ï¼Ÿ
    """,
    expected_output='æ¯å€‹è§’åº¦æä¾› 2-3 å€‹é‡é»ç™¼ç¾',
    agent=researcher
)

# ä»»å‹™2ï¼šæ’°å¯«
writing_task = Task(
    description="""
    æ ¹æ“šç ”ç©¶çµæœï¼Œæ’°å¯«ä¸€ç¯‡é—œæ–¼ã€Œæ°¸çºŒç™¼å±•èˆ‡ä¼æ¥­è²¬ä»»ã€çš„æ–‡ç« ã€‚
    
    æ–‡ç« çµæ§‹ï¼š
    1. å¼•è¨€ï¼ˆ1æ®µï¼‰
    2. ä¸»è¦å…§å®¹ï¼ˆ2-3æ®µï¼‰
    3. çµè«–ï¼ˆ1æ®µï¼‰
    """,
    expected_output='å®Œæ•´çš„æ–‡ç« ï¼Œç´„ 300-400 å­—',
    agent=writer
)

# ä»»å‹™3ï¼šç·¨è¼¯
editing_task = Task(
    description="""
    æª¢æŸ¥å’Œæ”¹å–„æ–‡ç« ï¼š
    1. ç¢ºä¿é‚è¼¯é€šé †
    2. ä¿®æ­£éŒ¯èª¤
    3. æå‡å¯è®€æ€§
    """,
    expected_output='ç·¨è¼¯å¾Œçš„æœ€çµ‚æ–‡ç« ',
    agent=editor
)

print("âœ… ä¸‰å€‹ä»»å‹™å®šç¾©å®Œæˆï¼")
```

```python
# çµ„æˆåœ˜éšŠä¸¦åŸ·è¡Œ
writing_crew = Crew(
    agents=[researcher, writer, editor],
    tasks=[research_task, writing_task, editing_task],
    verbose=True
)

# é–‹å§‹åŸ·è¡Œï¼
print("ğŸš€ é–‹å§‹åŸ·è¡Œå¯«ä½œæµç¨‹...\n")
final_article = writing_crew.kickoff()
```
