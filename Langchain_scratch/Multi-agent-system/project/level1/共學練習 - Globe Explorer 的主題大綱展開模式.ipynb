# 老師的探索記錄
- 這兩個生成的 outline 有什麼語意結構上的不同？
ChatGPT 的回覆：
兩個版本在語意結構上的確有顯著差異。版本一的層次性較為明顯，而版本二的結構則較為扁平。以下是對兩者的比較分析：

版本一（Globe Explorer 版）
```
建立一個專業的 LinkedIn 首頁

個人資訊
  - 姓名
  - 職稱
  - 地區
  - 個人簡介

專業背景
  - 教育經歷
    - 學校
    - 學位
    - 畢業年份
  - 工作經驗
    - 公司
    - 職位
    - 工作時間

技能與專長
  - 硬性技能
    - 程式語言
    - 軟體應用
    - 專業證照
  - 軟性技能
    - 溝通能力
    - 團隊合作
    - 問題解決

社群互動
  - 關注對象
    - 業界大咖
    - 同儕好友
    - 公司企業
  - 發表內容
    - 工作心得
    - 產業洞見
    - 生涯規劃

個人品牌
  - 個人標誌
    - 頭像
    - 背景圖
  - 個人網站
    - 作品集
    - 部落格
  - 社群影響力
    - 追蹤人數
    - 互動熱度
```
特點：

- 明確層次結構：第一層級為主要類別（如個人資訊、專業背景、技能與專長等），第二層級則為這些類別的子項目（如姓名、職稱等）。
- 層次分明：每一層次都能清晰地展示出所屬的子項目，使得結構更加層次化。
- 全面覆蓋：涵蓋了建立 LinkedIn 頁面所需的各個方面，詳細且全面。

版本二（STORM 長文寫作版）
```
# 建立一個專業的 LinkIn 首頁

## 個人資訊
- 添加專業頭像照片
- 更新個人標題和簡介
- 提供正確的聯絡資訊

## 經歷和學歷
- 詳細列出工作經驗
- 加入學歷背景
- 強調成就和貢獻

## 技能與專長
- 列出關鍵技能
- 獲得背書和推薦
- 添加認證或證書

## 專業照片和視覺元素
- 使用專業背景圖片
- 發佈專業形象照片
- 確保整體風格一致

## 網絡和互動
- 加入相關社群和組織
- 參與討論和留言
- 分享原創內容和文章
```

特點：

- 扁平結構：每個部分都是獨立的，不像版本一有明顯的主從關係。
- 直接指南：每個項目都是具體的建議或行動，例如「添加專業頭像照片」、「更新個人標題和簡介」。
- 簡潔明了：較少的層次使得大綱更為簡潔，適合快速閱讀和實施。

總結
- 層次性：版本一的結構更有層次感，便於逐層展開各個方面的詳細內容，適合需要深入了解的讀者。
- 直接性：版本二則更直接地提供具體建議和行動步驟，適合希望快速了解和操作的讀者。

總的來說，版本一在語意結構上更具有層次性，而版本二則更加扁平和直接。根据你的需求選擇適合的結構會更有幫助。

````python
# 取得 環境變數
from google.colab import userdata
from litellm import completion, acompletion
import json

OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')

topic = '建立一個專業的 LinkIn 首頁'
````

````python
def generate_outline(topic: str) -> str:
    """
    產生初版文章大綱。
    """

    system_prompt = """
    為一個特定主題撰寫文章大綱。下方是你寫作的格式：

    1. 使用“# 標題”來指示節的標題。
    2. 使用“## 標題”來指示小節的標題。
    3. 使用“-”符號來列出小節中的要點。
    4. 每個節和小節之間請確保有空行。
    5. 大綱應僅包含標題和要點，不應包含其他額外資訊。
    6. 請使用繁體中文。
    """

    user_prompt = """這裏是主題:\n\n主題:{topic}"""

    try:

        response = completion(
            api_key = OPENAI_API_KEY,
            model="gpt-3.5-turbo",
            messages=[{ "content": system_prompt,"role": "system"},{ "content": user_prompt.format(topic=topic),"role": "user"}],
            temperature=0.5,
            max_tokens=500,
        )
        response = response.choices[0].message.content

        return response
    except Exception as e:
        print(f"Error in translation: {e}")
````

````python
# 使用使用者指定的主題，產生主題大綱初版
outline = generate_outline(topic)
````

````python
print(outline) # generated outline
````
